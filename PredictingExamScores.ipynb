```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
# Sample data
data = {
    'Study Hours': [2, 3, 1],
    'Final Exam Score': [65, 75, 60]
}
```

```python
# Create a DataFrame
df = pd.DataFrame(data)
```

```python

# Features and target variable
X = df[['Study Hours']]  # Feature
y = df['Final Exam Score']  # Target variable
```

```python

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python

# Initialize and train the model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python

# Make predictions
y_pred = model.predict(X_test)
```

```python

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

```

```python

# Print the coefficients
print(f'Coefficient (m): {model.coef_[0]}')
print(f'Intercept (b): {model.intercept_}')
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

```python

# Plot the results
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, model.predict(X), color='red', linewidth=2, label='Fitted line')
plt.xlabel('Study Hours')
plt.ylabel('Final Exam Score')
plt.title('Study Hours vs Final Exam Score')
plt.legend()
plt.show()
```
