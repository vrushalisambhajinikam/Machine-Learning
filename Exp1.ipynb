```python
import numpy as np

# Create a 1D array
arr1 = np.array([1, 2, 3, 4, 5])

# Create a 2D array
arr2 = np.array([[1, 2, 3], [4, 5, 6]])

print("1D Array:", arr1)
print("2D Array:", arr2)
```

```python
#Array Operations
# Basic operations
arr_sum = arr1 + 2   # Add 2 to every element
arr_mult = arr1 * 3  # Multiply every element by 3

# Array math
arr_sqrt = np.sqrt(arr1)  # Square root of each element

print("Sum:", arr_sum)
print("Multiplied:", arr_mult)
print("Square root:", arr_sqrt)
```

 Introduction to Pandas

```python
#Creating a DataFrame:
import pandas as pd

# Creating a simple DataFrame
data = {'Name': ['John', 'Anna', 'Peter', 'Linda'],
        'Age': [28, 24, 35, 32],
        'City': ['New York', 'Paris', 'Berlin', 'London']}
df = pd.DataFrame(data)
print(df)
```

```python
#Basic DataFrame Operations:
# Accessing columns
print(df['Name'])

# Descriptive statistics
print(df.describe())

# Filtering
filtered_df = df[df['Age'] > 30]
print(filtered_df)
```

```python
#Reading a CSV File:
# Load a dataset from a CSV file
df = pd.read_csv('your_dataset.csv')
print(df.head())  # Show the first 5 rows
```

```python
#Step 4: Data Visualization with Matplotlib
#Plotting Simple Graphs:
import matplotlib.pyplot as plt

# Simple line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
#Bar Plot:
# Bar plot using pandas data
df.plot(kind='bar', x='Name', y='Age', color='green')
plt.show()
```

```python
#Scatter Plot:
# Scatter plot
plt.scatter(x, np.cos(x), color='red')
plt.title('Cosine Scatter')
plt.show()
```

```python
#Step 5: Combining the Libraries
# Generate random data using NumPy
data = np.random.randn(1000)

# Create a DataFrame from NumPy array using Pandas
df = pd.DataFrame(data, columns=['Random Numbers'])

# Plot a histogram of the data using Matplotlib
df['Random Numbers'].plot(kind='hist', bins=30, color='blue')
plt.title('Histogram of Random Numbers')
plt.show()
```
